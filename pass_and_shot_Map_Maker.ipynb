```python
import pandas as pd
import matplotlib.pyplot as plt
!pip install highlight_text
from highlight_text import fig_text
import matplotlib as mpl
!pip install mplsoccer
from mplsoccer.pitch import Pitch
from mplsoccer.pitch import VerticalPitch

```

```python
CSVdir = input('Enter File Directory: ')
```

```python
data = pd.read_csv(CSVdir)


print(data)
```

```python
title = input('Enter Shot Map Title: ')
```

#SHOT MAP


```python
pitch = Pitch(pitch_type='opta',
              pitch_length=105, pitch_width=68, pitch_color='#000236', line_color='White')
fig, ax = pitch.draw()

plt.scatter(data['X'],data['Y'], c='#21fffb', s=100)

fig.suptitle(title, fontsize=20, color="White", fontweight='bold')
fig.set_size_inches(15, 7)
```

#PASS MAP

```python
pitch = Pitch(pitch_type='opta', orientation = 'horizontal',
              pitch_length=105, pitch_width=68, pitch_color='#000236', line_color='White')

fig, ax = pitch.draw()

plt.plot([data["X"], data["X2"]], [data["Y"], data["Y2"]], color="#21fffb", linewidth= 1.0)

plt.scatter(data["X"],data["Y"], s=10,  color="#21fffb")

plt.show()

fig.suptitle(title, fontsize=20, color="White", fontweight='bold')

fig.suptitle(title, fontsize=20, color="White", fontweight='bold')


fig.set_size_inches(15, 7)
```
